```python
mexico = gpd.read_file("/content/mexican-states/mexican-states.shp")

scope = ["Tamaulipas", "Nuevo León", "Coahuila de Zaragoza", "San Luis Potosí",
         "Aguascalientes", "Zacatecas", "Chihuahua", "Durango",
         "Sonora", "Baja California", "Baja California Sur",
         "Sinaloa", "Jalisco", "Colima"]

arca = mexico.query(f'name == @scope')
arca
```

```python
sns.set(style="white", font_scale = 0.5, color_codes=True)

fig, ax = plt.subplots(1, 1, figsize = (15, 15))

arca.plot(ax = ax, color = "lightgrey", edgecolor = "white")

plt.axis('off')
plt.title("Cobertura por AGEB") # 

plt.show()
```
